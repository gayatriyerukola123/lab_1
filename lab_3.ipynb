```python
import pandas as pd
import numpy as np
import statistics

file_path="C:/Users/year3/Downloads/Lab Session Data.xlsx"
sheet_name='IRCTC Stock Price' 
data=pd.read_excel(file_path,sheet_name)
print(data)
price=data['Price']
price_list=price.tolist()
mean=statistics.mean(price_list)
var=statistics.variance(price_list)
print(f"the mean is {mean} and variance is {var}")
```

```python
wed_data=data[data['Day']=='Wed']
price=wed_data['Price']
price_list=price.tolist()
print(price_list)
mean=statistics.mean(price_list)
var=statistics.variance(price_list)
print(f"the mean is {mean} and variance is {var}")
```

```python
april_data=data[data['Month'] == 'Apr']
price=april_data['Price']
price_list=price.tolist()
print(price_list)
mean=statistics.mean(price_list)
var=statistics.variance(price_list)
print(f"the mean is {mean} and variance is {var}")
```

```python
losses=data['Chg%'].apply(lambda x:x<0)
prob=(data[losses].shape[0])/data.shape[0]
print(prob)
```
